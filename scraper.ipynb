```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
import re
import pandas as pd
import matplotlib.pyplot as plt
import time
```

```python
browser = Browser('chrome')
base_url = 'https://www.facebook.com/marketplace/austin/search?exact=false'

radius = 400
min_price = 0
max_price = 20000
days_listed = 30
min_mileage = 0
max_mileage = 300000
min_year = 2001
max_year = 2012
transmission = 'automatic'
make = 'Honda'
model = 'Element'

url = f"{base_url}minPrice={min_price}&maxPrice={max_price}&daysListed={days_listed}&minMileage={min_mileage}&maxMileage={max_mileage}&minYear={min_year}&maxYear={max_year}&transmissiontype={transmission}&query={make}{model}&radius={radius}&exact=false"
```

```python
browser.visit(url)
```

```python
 if browser.is_element_present_by_css('div[aria-label="Close"]', wait_time=10):
    # Click on the element once it's found
    browser.find_by_css('div[aria-label="Close"]').first.click()
```

```python
scroll_count = 4
scroll_delay = 2

for i in range(scroll_count):
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(scroll_delay)
```
